```python
import numpy as np
import scipy.special
import pandas as pd

#Import Sci Kit Learn KMeans
from sklearn.cluster import KMeans

#Related Bokeh libreries
import bokeh.io
import bokeh.plotting
from bokeh.plotting.figure import figure

# Enable viewing Bokeh plots in the notebook
bokeh.io.output_notebook()
```

```python
test_array = np.array([[3, 2], [2, 4], [8, 9],[10,8], [10, 9], [2, 5], [3, 4], [10, 8], [3, 5], [7, 8], [10, 7], [9, 7]])
```

```python
kmeans = KMeans(n_clusters=2).fit(test_array)
print(kmeans.labels_)
print(kmeans.predict([[3, 4], [8, 9]]))
print(kmeans.cluster_centers_)
```

```python
f = figure(plot_width=300,plot_height=300)
f.circle(x=test_array[:,0],y=test_array[:,1], size=7)
bokeh.io.show(f)
```

```python

f.circle(x=kmeans.cluster_centers_[:,0],y=kmeans.cluster_centers_[:,1], size=15, color="orangered", fill_alpha=0.7)
bokeh.io.show(f)
```
